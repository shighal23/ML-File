## 2. Understanding KNN
k-Nearest Neighbors (KNN) is a simple, non-parametric, lazy learning algorithm used for classification and regression. It classifies a data point based on the majority class of its 'k' nearest neighbors.

## 3. Dataset Description
We are using the **Breast Cancer Wisconsin (Original) Data Set**. The goal is to predict whether a tumor is benign or malignant based on various attributes like clump thickness, cell size, etc.

## 11. Model Training
We train the KNN classifier with k=3.

```python
X = df.drop(["Class"], axis=1)

y = df["Class"]
```

## 12. Prediction
Predict the results on the test set.

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## 14. Rebuilding Model with k=7
We retrain the model with k=7 to see if accuracy improves.

```python
# impute missing values in X_train and X_test with respective column median in X_train

for df1 in [X_train, X_test]:
    for col in X_train.columns:
        col_median = X_train[col].median()
        df1[col].fillna(col_median, inplace=True)
```

## 16. Classification Metrics
Print the classification report containing precision, recall, and f1-score.

```python
# import KNeighbors ClaSSifier from sklearn
from sklearn.neighbors import KNeighborsClassifier


# instantiate the model
knn = KNeighborsClassifier(n_neighbors=3)


# fit the model to the training set
knn.fit(X_train, y_train)
```

```python
# print classification accuracy

classification_accuracy = (TP + TN) / float(TP + TN + FP + FN)

print("Classification accuracy : {0:0.4f}".format(classification_accuracy))
```

```python
# print classification error

classification_error = (FP + FN) / float(TP + TN + FP + FN)

print("Classification error : {0:0.4f}".format(classification_error))
```

```python
# print precision score

precision = TP / float(TP + FP)


print("Precision : {0:0.4f}".format(precision))
```

```python
recall = TP / float(TP + FN)

print("Recall or Sensitivity : {0:0.4f}".format(recall))
```

```python
true_positive_rate = TP / float(TP + FN)


print("True Positive Rate : {0:0.4f}".format(true_positive_rate))
```

```python
false_positive_rate = FP / float(FP + TN)


print("False Positive Rate : {0:0.4f}".format(false_positive_rate))
```

```python
specificity = TN / (TN + FP)

print("Specificity : {0:0.4f}".format(specificity))
```
