## 5. Histograms
Histograms show the distribution of a single numerical variable. They help us see the frequency of data points in different ranges (bins).

## 8. Advanced Visualizations
In this section, we will explore more advanced plots like Point Plots, Donut Charts, and Box Plots using the 'flights' and 'tips' datasets.

```python
df.head(12)
```

**Point Plot**

```python
# plot means estimates and confidence for categorical data.
```

plt.title("Average passengers as per year")

sns.pointplot(x='year',data=df,y='passengers',hue='month')
plt.show()

```python
df.groupby(["month", "year"])["passengers"].sum().astype("int")
```

**Donut Chart**

```python
colors = sns.color_palette("Set3", 12)
plt.figure(figsize=(6, 6))
plt.pie(
    df.groupby("year")["passengers"].mean().astype("int"),
    labels=df.year.unique(),
    colors=colors,
    autopct="%1.1f%%",
    startangle=220,
    wedgeprops={"width": 0.3},
)

plt.title("Tips distribution Years wise")
plt.show()
```

**Cat Plot**

```python
# categorical "plot" function.
```

```python
sns.catplot(x="year", y="passengers", data=df, hue="month", kind="swarm")
plt.figure(figsize=(12, 6))
plt.show()
```

**Box plot**

```python
sns.categorical.boxenplot(df.passengers, fill=True)
plt.show()
```

```python
# Group by year to get total passengers per year
yearly_passengers = df.groupby("year", as_index=False)["passengers"].sum()

# top 3 year
top_3 = yearly_passengers.sort_values(by="passengers", ascending=False).head(3)

# Bottom 3 year
bottom_3 = yearly_passengers.sort_values(by="passengers", ascending=True).head(3)

# combine
top_bottom_years = pd.concat([top_3, bottom_3]).reset_index(drop=True)
plt.figure(figsize=[12, 6])
sns.barplot(
    x="year", y="passengers", data=top_bottom_years, palette="Set2", errorbar=None
)
plt.title("Top 3 and Bottom 3 Years by total passengers")
plt.xticks(rotation=45)
plt.show()
```
